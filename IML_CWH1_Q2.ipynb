## Array creation
>Create the followings to numpy array:
* A: a 3\*3 array like this:
$$
A = \left(\begin{array}{cc} 
10 & 20 & 30\\
5 & 10 & 15\\
1 & 5 & 8
\end{array}\right)
$$ 
* B: a 3\*3\*4 that filled with the numbers start from 15 and have steps of 5 to the end. The array should be filled first along the first dimension then the second dimension and at last the third dimension. like this:
$$
B[:, :, 0] = \left(\begin{array}{cc} 
15 & 30 & 45\\
20 & 35 & 50\\
25 & 40 & 55
\end{array}\right)
$$ 

```python
### START CODE HERE ###
A = np.array([[10,20,30],[5,10,15],[1,5,8]])
b = np.arange(15,15 + 3*3*4*5,5)
B = np.transpose(b.reshape(4,3,3))
print('A =',A)
print('B[:,:,0] =',B[:,:,0])
### END CODE HERE ###
```

## Replacement
>Replace the last slice of array B (B[:, :, 3]) with values of A. Then print the result.

```python
### START CODE HERE ###
B[:,:,3] = A.copy()
print('B[:,:,3] =',B[:,:,3])
### END CODE HERE ###
```

## Axis Swap 
>Swap the first(B[:, :, 0]) and last(B[:, :, 3]) slice of the array B **only with one line of code**. Then print the result.

```python
### START CODE HERE ###
B[:,:,[0,3]] = B[:,:,[3,0]]
print('B[:,:,0] =',B[:,:,0])
print('B[:,:,3] =',B[:,:,3])
### END CODE HERE ###
```

## Transposing 
>Transpose the array B elements in each slice. Then print the result.

```python
### START CODE HERE ###

B[:,:,0]=B[:,:,0].T
print('B[:,:,0] =',B[:,:,0])
B[:,:,1]=B[:,:,1].T
print('B[:,:,1] =',B[:,:,1])
B[:,:,2]=B[:,:,2].T
print('B[:,:,2] =',B[:,:,2])
B[:,:,3]=B[:,:,3].T
print('B[:,:,3] =',B[:,:,3])

### END CODE HERE ###
```

## Matrix Multiplication
>Calculate the matrix multiplication of all slices of the array B. 
(B[:, :, 0] \* B[:, :, 1] \* B[:, :, 2] \* B[:, :, 3]). Then print the result.

>**Note**: the result should be a 3\*3 array. 

```python
### START CODE HERE ###
multiply = np.matmul(B[:,:,0],B[:,:,1])
multiply = np.matmul(multiply,B[:,:,2])
multiply = np.matmul(multiply,B[:,:,3])
print(multiply)

### END CODE HERE ###
```

## Determinant
>Calculate the determinant of the array A and print it.

```python
### START CODE HERE ###
determinant = np.linalg.det(A)
print('determinant of A is',determinant)
### END CODE HERE ###
```

## Actors/Actresses Activity
>For each Actors/Actresses caculate the number of movies in the dataframe, then sort the actors/actresses first by their number of movies in descending order and then by their name in ascending order. 

```python
### START CODE HERE ###

actors=[]
for i in range(len(df['Actors'])):
    x = list(df['Actors'].iloc[i].split(', '))
    for k in range(len(x)):
        y = list(x[k].split(','))
        for j in range(len(y)):
            actors.append(y[j]) 
print("Actors/actresses sorted by their number of movies\n")
numberOfMovies = pd.Series(actors).value_counts().sort_index().sort_values(ascending = False)
print(numberOfMovies)

print("\nActors/actresses sorted by their name\n")
name = pd.Series(actors).value_counts().sort_index(ascending = True)
print(name)
### END CODE HERE ###
```

## Correlations
>Calculate the correlation between rating, votes and metascore of the movies.
Is there any meaningful relation between these features? 

```python
### START CODE HERE ###
corr1 = df["Rating"].corr(df["Votes"])
corr2 = df["Rating"].corr(df["Metascore"])
corr3 = df["Votes"].corr(df["Metascore"])
print('The correlation between rating and votes is',corr1)
print('The correlation between rating and metascore is',corr2)
print('The correlation between votes and metascore is',corr3)
### END CODE HERE ###
```

As we know the correlation is a number between -1 and 1. We can see that the correlations are positive so every two feature have a direct relation with each other and as the correlation gets higher, the realtion gets stronger. For example the correlation between the rating and the Metascore is 0.63 which implies that a higher rating can lead to a higher Metascore.

## Plot
>You must plot the cos(x) curve. plot 20 cycle of the curve. the frequency of first half of curve is 50 Hz and frequency of second lalf is 200 Hz. 
> set the x_label, y_label and title for your plot.

```python
### START CODE HERE ###
x = np.arange(0,10*(1/50)+10*(1/200),0.0001)
y = []
for i in range(0,len(x)):
  if x[i]<10*(1/50):
    z = np.cos(2*np.pi*50*x[i])
  else:
    z = np.cos(2*np.pi*200*x[i])
  y.append(z)
fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x,y)
plt.title('Cos(x): 10 cycle with f=50 Hz and 10 cycle with f=200 Hz')
### END CODE HERE ###
```
